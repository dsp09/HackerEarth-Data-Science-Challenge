# HackerEarth Data Science Challenge 2019

```python
import pandas as pd
```

```python
train = pd.read_csv("dataset/train.csv")
test = pd.read_csv("dataset/test.csv")

print(train.shape, test.shape)
```

```python
train.head()
```

```python
test.tail()
```

```python
reference = {"h0": 0, "h1": 1, "h2": 2, "h3": 3, "h4": 4, "h5": 5}
```

```python
train_data = train.loc[:, "para_0":"para_50"]
train_output = train["Result"].map(reference)

test_data = test.loc[:, "para_0":]

print(train_data)
print(train_data.shape, test_data.shape)
```

### data splitting

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(train_data, train_output, test_size=0.33, random_state=42)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
print(X_train.head(), X_test.head(), y_train.head(), y_test.head())
```

```python
X_train.reset_index(inplace=True)
X_test.reset_index(inplace=True)
# y_train.reset_index(inplace=True)
# y_test.reset_index(inplace=True)
```

```python
print(X_train.head())
```

```python
from sklearn.metrics import f1_score
```

```python
def pred_f1_score(X_train, X_test, y_train, y_test, clf):
    clf.fit(X_train,y_train)
    predTrain = pd.Series(clf.predict(X_train))
    predTest = pd.Series(clf.predict(X_test))
    f1_train = f1_score(y_train, predTrain, average='micro')
    f1_test = f1_score(y_test, predTest, average='micro')
    return (f1_train, f1_test)
```

### Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clfRandom = RandomForestClassifier(max_depth=15, random_state=0)
```

```python
pred_f1_score(X_train, X_test, y_train, y_test, clfRandom)
```

### XGBoost

```python
from xgboost import XGBClassifier
```

```python
clfXGB = XGBClassifier()
```

```python
pred_f1_score(X_train, X_test, y_train, y_test, clfXGB)
```

### Support Vector Classifier

```python
from sklearn.svm import SVC
```

```python
clfSVC = SVC(kernel='linear')
```

```python
pred_f1_score(X_train, X_test, y_train, y_test, clfSVC)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
clfLogreg = LogisticRegression()
```

```python
pred_f1_score(X_train, X_test, y_train, y_test, clfLogreg)
```
